```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 제가 만든 py 파일을 불러옵니다.

import ti_matrix as tm
```

```python
# ohlcv가 포함된 DataFrame을 다운로드 받습니다.

import FinanceDataReader as fdr
df = fdr.DataReader('005930','2005-1-1')
df
```

```python
# 각 기술적지표의 window를 list로 설정합니다
windows = np.arange(20,125,10)
windows
```

```python
matrix = tm.get_all_signal(df, windows).dropna()
matrix
```

```python
# 저장

matrix.to_csv('c:ti_matrix.csv')
```

그림으로 보기

```python
close = df.close
```

```python
for i in matrix.columns:
    sc = matrix[i].copy()
    plt.figure(figsize=(10,3))
    plt.plot(close.loc[sc.index], linewidth=0.5,alpha=0.6)
    plt.scatter(close.loc[sc.index].index, close.loc[sc.index], c=sc,cmap='gray_r', alpha=1, vmin=0,vmax=1)
    plt.colorbar()
    plt.title('{}'.format(i))
    plt.show()
```

```python
import seaborn as sns;sns.set()

plt.figure(figsize=(40,40))
sns.heatmap(matrix.corr(), cmap='gray_r')
plt.title('Correlation of binary tech. ind.', )
plt.show()
```
